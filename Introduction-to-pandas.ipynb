### Fill missing values using scikit-learn

```python
car_sales_missing = pd.read_csv('car-sales-extended-missing-data.csv')
```

```python
car_sales_missing.isna().sum()
```

```python
car_sales_missing.dropna(subset=['Price'],inplace=True)
car_sales_missing.isna().sum()
```

```python
# split into X and y
X = car_sales_missing.drop('Price',axis=1)
y = car_sales_missing['Price']
```

```python
X
```

```python
# fill missing values with scikit-learn
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

# fill the catogorical values with 'missing' and numerical values with mean
cat_imputer = SimpleImputer(strategy='constant',fill_value='missing')
door_imputer = SimpleImputer(strategy='constant',fill_value=4)
num_imputer = SimpleImputer(strategy='mean')

# define coloumn
cat_features = ['Make','Colour']
door_features = ['Doors']
num_features = ['Odometer (KM)']

# Create an imputer(something that fill the missing values)
imputer = ColumnTransformer([('cat_imputer',cat_imputer,cat_features),
                             ('door_imputer',door_imputer,door_features),
                             ('num_imputer',num_imputer,num_features)])

# Transform the data
filled_X = imputer.fit_transform(X)
filled_X
```

```python
car_sales_filled = pd.DataFrame(filled_X,
                                columns=['Make','Colour','Doors','Odometer (KM)'])
```

```python
car_sales_filled
```

```python
car_sales_filled.isna().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
catogorical_feature = ['Make','Colour','Doors']
one_hot = OneHotEncoder()
transformer = ColumnTransformer([('one_hot',one_hot,catogorical_feature)],remainder='passthrough')
transformed_x = transformer.fit_transform(car_sales_filled)
transformed_x
```

```python
np.random.seed(42)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(transformed_x,y,test_size=0.2)
```

```python
model = RandomForestRegressor()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
